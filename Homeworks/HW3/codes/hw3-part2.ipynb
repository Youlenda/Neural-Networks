```
import numpy as np
```

```
s = np.matrix([1, 1, 1, -1])
t = np.matrix([1, 1, 1, -1])
# s.shape
```

```
i = np.identity(4)
# i
```

```
w = s.T * s - i
# w.shape
w
```

```
index = np.random.choice(4, 1, replace = False)
print(index)
```

```
for i in range (0,1):
  if s[0,index[i]] == 1:
    s[0,index[i]] = -1
  else:
    s[0,index[i]] = 1
s
```

```
out = s * w

for i in range (0,4):
  if out[0,i] > 0:
     out[0,i] = 1
  else:
     out[0,i] = -1

print('Output is',out,'.')
print('Target is',t,'.')
z = np.count_nonzero(t != out)
print("Hamming distance between output and target is", z,'.')
```
