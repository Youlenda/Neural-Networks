```
identity = np.identity(64)
w = np.dot(sample.T, sample) - 10 * identity
w = w/250
```

```
def af(x,y):
  if x > 0:
    return 1
  elif x < 0:
    return -1
  else:
    return y
```

```
y = np.array(n_sample)
target = np.array(n_sample)

y_in = np.zeros((10,64))
```

```
flag = True
for i in range (0, 10):
  flag = True
  while (flag):
    ro = np.random.permutation(64) #random order
    for j in range (0, 64):
      sum_ = 0
      for k in range (0, 64):
        sum_ = sum_ + y[i, k] * w[k, ro[j]] 
      y_in[i, ro[j]] = sum_ + y[i, ro[j]]
      y[i, ro[j]] = af(y_in[i, ro[j]], y[i, ro[j]])
      if target[i, ro[j]] == y[i, ro[j]]:
        flag = False
```
